```python
import numpy as np
import matplotlib.pyplot as plt
import timeit
```

```python
def bisection(f, a0, b0, eps=1E-3):
    a, b = a0, b0
    while b - a > 2*eps:
        x = (a + b)/2
        if f(a)*f(x) < 0:
            b = x
        else:
            a = x
    return x

def regula_falsi(f, a0, b0, delta=1E-3):
    a, b = a0, b0
    xnew = (a+b)/2
    xold = a
    while abs(xnew - xold) > delta:
        xold = xnew
        xnew = a - f(a)*(b-a)/(f(b)-f(a))
        if f(a)*f(xnew) < 0:
            b = xnew
        else:
            a = xnew
    return xnew

def plot_result(f, a0, b0, x, title):
    plt.title(title)
    plt.plot(x, f(x), "bo")                                               # koÅ™en
    plt.plot(np.linspace(a0, b0, 100), f(np.linspace(a0, b0, 100)), "r")  # function
    plt.axhline(y=0)
    plt.axvline(x=0)
    plt.show()
```

```python
function_1 = lambda x: x**2 + x
function_2 = lambda x: np.exp(x) + 4*x
function_3 = lambda x: np.cos(x) + 0.5

hrana_1, hrana_2= -4, 4

plot_result(function_1, hrana_1, hrana_2, bisection(function_1, hrana_1, hrana_2), "Bisection of x^2 + x")
print(f"Bisection time for 100000 repeates: {timeit.timeit(lambda: bisection(function_1, hrana_1, hrana_2), number=100000)}/s")
plot_result(function_1, hrana_1, hrana_2, regula_falsi(function_1, hrana_1, hrana_2), "Regula flasi of x^2 + x")
print(f"Refula falsi time for 100000 repeates: {timeit.timeit(lambda: regula_falsi(function_1, hrana_1, hrana_2), number=100000)}/s")
```

```python
hrana_1, hrana_2= -4, 4

plot_result(function_2, hrana_1, hrana_2, bisection(function_2, hrana_1, hrana_2), "Bisection of e^x + 4*x")
print(f"Bisection time for 100000 repeates: {timeit.timeit(lambda: bisection(function_2, hrana_1, hrana_2), number=100000)}/s")
plot_result(function_2, hrana_1, hrana_2, regula_falsi(function_2, hrana_1, hrana_2), "Regula flasi of e^x + 4*x")
print(f"Regula falsi time for 100000 repeates: {timeit.timeit(lambda: regula_falsi(function_2, hrana_1, hrana_2), number=100000)}/s")
```

```python
hrana_1, hrana_2= -2*np.pi, 2*np.pi

plot_result(function_3, hrana_1, hrana_2, bisection(function_3, hrana_1, hrana_2), "Bisection of cos(x) + 0.5")
print(f"Bisection time for 100000 repeates: {timeit.timeit(lambda: bisection(function_3, hrana_1, hrana_2), number=100000)}/s")
plot_result(function_3, hrana_1, hrana_2, regula_falsi(function_3, hrana_1, hrana_2), "Regula flasi of cos(x) + 0.5")
print(f"Refula falsi time for 100000 repeates: {timeit.timeit(lambda: regula_falsi(function_3, hrana_1, hrana_2), number=100000)}/s")
```
