```python
import numpy as np
import random as rd
import timeit
import matplotlib.pyplot as plt
```

```python
def direct_method(A, B):
    matrix_A = np.array(A)
    matrix_B = np.array(B)
    x = np.linalg.solve(A, B)
    return x

def jacobi(A, B):
    matrix_A = np.array(A)
    matrix_B = np.array(B)
    x = np.ones(len(A))
    D = np.diag(matrix_A)
    L = np.tril(matrix_A, k = -1)
    U = np.triu(matrix_A, k = 1)
    for i in range(2, 7):
        x = (matrix_B - np.matmul((L + U),x))/D
    return x

def get_matrix(dimention: int):
    matrix = [[rd.randint(1, 10) for j in range(dimention)] for i in range(dimention)]
    matrix_eq = [rd.randint(1, 10) for i in range(dimention)]
    return matrix, matrix_eq

cas_list_dm = []
cas_list_it = []
velikost = []

for start in range(2, 50, 2):
    A, B = get_matrix(start)
    cas_dm = timeit.timeit(lambda: direct_method(A, B), number= 1)
    cas_it = timeit.timeit(lambda: jacobi(A, B), number= 1)
    if cas_it < cas_dm:
        print(f"Ted je {start} kolo!")
    cas_list_dm.append(cas_dm)
    cas_list_it.append(cas_it)
    velikost.append(start)
    
plt.plot(velikost, cas_list_dm)
plt.plot(velikost, cas_list_it)
plt.xlabel("Velikost matice", color="blue")
plt.ylabel("Čas potřebný na vypočtení", color="blue")
plt.title("Iterační VS přímá metoda", color="blue")
```
