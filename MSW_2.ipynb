```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
games = pd.read_csv("vgsales.csv")
```

```python
#seřadění a zkrácení aby se názvy vešli na graf
games_sorted = games.sort_values(by="Global_Sales", ascending=False)
games_set = games_sorted.head(15)
```

```python
#vytvoření grafu
sales_plot = sns.barplot(x = 'Global_Sales', y = 'Name', data = games_set)
plt.ylabel("Titles of Games")
plt.xlabel("Sales in Millions")
plt.title("Top 15 Worldwide sales", fontsize=30, color="darkblue")
```

```python
#slicing and using sum to get number of sold copies on eath continent
sales_by_continent = games_sorted[["Name", "NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales", "Global_Sales"]]
sales_by_continent_sum = sales_by_continent.sum()
regions = ["North America", "Europe", "Japan", "Other"]
plt.pie(sales_by_continent_sum[1:5], labels=regions, autopct='%.1f%%', wedgeprops={'linewidth': 2.0, 'edgecolor': 'black'}, textprops={'size': 'x-large'})
plt.title("Regions With Most Sales", fontsize= 25)
```

```python
#games["Year"].fillna(0).astype(int)
plt.figure(figsize=(20,10))
sns.countplot(games, x=games["Year"])
plt.xlabel("Year", fontsize= 20, color="blue")
plt.ylabel("Number of releases", fontsize= 20, color="blue")
plt.title("Number Of Games Released Each Year", fontsize= 40, color="blue")
p = plt.xticks(rotation = 45)
```

```python
games_2011 = games[games["Year"] >= 2011]
sns.lmplot(data= games_2011, x="Year", y="Global_Sales", fit_reg=False, hue="Platform", height= 10.0, aspect=12.0/10.0)
plt.ylim(-1,25)
plt.xlim(2010,2016.5)
plt.title("Distibution Of Games And Sales In Between Year 2011 and 2016", fontsize= 30, color="darkblue")
plt.ylabel("Global Sales", fontsize=20, color="darkblue")
plt.xlabel("Release Year", fontsize=20, color="darkblue")
```

```python
sns.lmplot(
    data=games, x="Year", y="Global_Sales", col="Platform", hue="Platform", fit_reg=False,
    col_wrap=4, palette="muted", ci=None,
    height=4, scatter_kws={"s": 50, "alpha": 1}
)
```
