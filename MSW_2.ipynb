```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
games = pd.read_csv("vgsales.csv")
```

```python
#seřadění a zkrácení aby se názvy vešli na graf
games_sorted = games.sort_values(by="Global_Sales", ascending=False)
games_set = games_sorted.head(15)
```

```python
#vytvoření grafu
sales_plot = sns.barplot(x = 'Global_Sales', y = 'Name', data = games_set)
plt.ylabel("Titles of Games")
plt.xlabel("Sales in Millions")
plt.title("Top 15 Worldwide sales", fontsize=30, color="darkblue")
```

```python
#slicing and using sum to get number of sold copies on eath continent
sales_by_continent = games_sorted[["Name", "NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales", "Global_Sales"]]
sales_by_continent_sum = sales_by_continent.sum()
regions = ["North America", "Europe", "Japan", "Other"]
plt.pie(sales_by_continent_sum[1:5], labels=regions, autopct='%.1f%%', wedgeprops={'linewidth': 2.0, 'edgecolor': 'black'}, textprops={'size': 'x-large'})
plt.title("Regions With Most Sales", fontsize= 25)
```
